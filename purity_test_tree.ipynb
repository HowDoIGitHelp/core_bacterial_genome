```python
import pandas
import numpy
raw = pandas.DataFrame.from_csv('P+N_DATA.csv')
train=raw.transpose()
train=raw.transpose()
for col in train.columns.values:
    for i in train.index.values:
        if(train[col][i]>0):
            train.set_value(i,col,1)
X=train.iloc[0:262, 0:4346]
Y=train['YCOMPLETE']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
numpy.arange(0.1,0.,0.01)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score


f1_score_list=[]
for impurity in numpy.arange(0.0,1,0.01):
    clf = RandomForestClassifier(n_estimators=1,min_impurity_split=impurity)
    clf = clf.fit(X_train, y_train)
    predictions=numpy.array(clf.predict(X_test))
    f1_score_list.append(f1_score(y_test,predictions))
f1_score_list
```

```python
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)

plt.plot(numpy.arange(0.0,1,0.01),f1_score_list)
plt.xlabel('minimum impurity')
plt.ylabel('f_score')
plt.show()
```
