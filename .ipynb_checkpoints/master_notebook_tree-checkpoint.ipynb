```python
import pandas
raw = pandas.DataFrame.from_csv('P+N_Train.csv')
```

```python
raw
```

```python
train=raw.transpose()
```

```python
train=raw.transpose()
for col in train.columns.values:
    for i in train.index.values:
        if(train[col][i]>0):
            train.set_value(i,col,1)
train
```

```python
X=train.iloc[0:197, 0:4346]
X
```

```python
Y=train['YCOMPLETE']
Y
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

clf = RandomForestClassifier(n_estimators=10,min_impurity_split=0.1)
clf = clf.fit(X, Y)
```

```python
from sklearn import tree
i = 0
for e in clf.estimators_ :
    with open("tree"+str(i)+".dot", 'w') as f:
        f = tree.export_graphviz(e, out_file=f, feature_names=list(train.columns.values), filled=True, rounded=True, special_characters=True)
    i = i + 1
```

```python
raw_test = pandas.DataFrame.from_csv('P+N_Test.csv')
test=raw_test.transpose()
for col in test.columns.values:
    for i in test.index.values:
        if(test[col][i]>0):
            test.set_value(i,col,1)
X_test=test.iloc[0:197, 0:4346]
Y_test=test['YCOMPLETE']
test
```

```python
actual=pandas.DataFrame(Y_test.tolist(),columns=['Prediction'],index=list(range(65)))
actual
```

```python
import numpy
opinions=numpy.empty([65,10])
for e in clf.estimators_ :
    print(numpy.array(e.predict(X_test)))
```

```python
import numpy
opinion_list=list()
for e in clf.estimators_ :
    opinion_list.append(numpy.array(e.predict(X_test)))
opinions=numpy.array(opinion_list)
opinions
```

```python
predictions=pandas.DataFrame(opinions.transpose(),columns=list(range(10)),index=list(range(65)))
result = pandas.concat([predictions,actual],axis=1,join='inner')
result
```
