```python
import pandas
raw = pandas.DataFrame.from_csv('P+N_DATA.csv')
```

```python
raw
```

```python
train=raw.transpose()
```

```python
train=raw.transpose()
for col in train.columns.values:
    for i in train.index.values:
        if(train[col][i]>0):
            train.set_value(i,col,1)
train
```

```python
cog_distributions={}
for header in list(train.columns.values):
    distribution=0
    for index in range(261):
        if(train.get_value(str(index),header)>0):
            distribution=distribution+1
    cog_distributions[header]=distribution
cog_distributions
```

```python
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)

plt.hist(list(cog_distributions.values()),64)
plt.xlabel('COG occurences')
plt.ylabel('distribution')
plt.grid(True)
plt.show()
```

```python
train.to_csv('booleanized_data.csv', sep=',', encoding='utf-8')
```

```python
X=train.iloc[0:262, 0:4346]
X
```

```python
Y=train['YCOMPLETE']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

clf = RandomForestClassifier(n_estimators=1,min_impurity_split=0.1)
clf = clf.fit(X_train, y_train)
```

```python
from sklearn import tree
i = 0
for e in clf.estimators_ :
    with open("tree"+str(i)+".dot", 'w') as f:
        f = tree.export_graphviz(e, out_file=f, feature_names=list(train.columns.values), filled=True, rounded=True, special_characters=True)
    i = i + 1
```

```python
clf.predict(X_test)
```

```python
actual=pandas.DataFrame(y_test.tolist(),columns=['Prediction'],index=list(range(y_test.shape[0])))
actual
y_test.shape
```

```python
import numpy
opinion_list=list()
for e in clf.estimators_ :
    opinion_list.append(numpy.array(e.predict(X_test)))
opinion_list.append(numpy.array(clf.predict(X_test)))
opinions=numpy.array(opinion_list)
```

```python
predictions=pandas.DataFrame(opinions.transpose(),columns=list(range(2)),index=list(range(y_test.shape[0])))
result = pandas.concat([predictions,actual],axis=1,join='inner')
result
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,result[1])
```
