```python
import pandas
import numpy
raw = pandas.DataFrame.from_csv('P+N_DATA.csv')
train=raw.transpose()
train=raw.transpose()
for col in train.columns.values:
    for i in train.index.values:
        if(train[col][i]>0):
            train.set_value(i,col,1)
X=train.iloc[0:262, 0:4346]
Y=train['YCOMPLETE']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score



f1_score_list=[]
for neurons in numpy.arange(1,4000,80):
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(2, neurons), random_state=1)
    clf = clf.fit(X_train, y_train)
    predictions=numpy.array(clf.predict(X_test))
    f1_score_list.append(f1_score(y_test,predictions))
```

```python
f1_score_list
```

```python
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (20, 10)

plt.plot(numpy.arange(1,4000,80),f1_score_list)
plt.xlabel('layers')
plt.ylabel('f_score')
plt.show()
```
